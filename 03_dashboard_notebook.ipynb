```python
df = spark.table("crime_final_cluster.default.clean_crime")
df.show(5)
```

```python
## Crime by Area_Name
from pyspark.sql.functions import count

dash_crimes_by_area = (
    df.groupBy("area_name")
      .agg(count("*").alias("crime_count"))
)

dash_crimes_by_area.write.mode("overwrite").saveAsTable("crime_final_cluster.default.dash_crimes_by_area")
```

```python
## Crime by Year
dash_crimes_by_year = (
    df.groupBy("occ_year")
      .agg(count("*").alias("crime_count"))
)

dash_crimes_by_year.write.mode("overwrite").saveAsTable("crime_final_cluster.default.dash_crimes_by_year")

```

```python
from pyspark.sql.functions import count

df = spark.table("crime_final_cluster.default.clean_crime")

dash_crimes_by_age_group = (
    df.groupBy("age_group")
      .agg(
          count("*").alias("crime_count")
      )
)

dash_crimes_by_age_group.write.mode("overwrite").saveAsTable("crime_final_cluster.default.dash_crimes_by_age_group")
```

```python
from pyspark.sql.functions import count

dash_top_crime_categories = (
    df.groupBy("crm_cd_desc")
      .agg(count("*").alias("crime_count"))
      .orderBy("crime_count", ascending=False)
      .limit(10)
)

dash_top_crime_categories.write.mode("overwrite").saveAsTable(
    "crime_final_cluster.default.dash_top_crime_categories"
)
```
