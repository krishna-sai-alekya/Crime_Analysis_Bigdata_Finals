```python
## Load Clean Table 
df = spark.table("crime_final_cluster.default.clean_crime")
df.show(5)
df.printSchema()
```

```python
## Crime per Year
from pyspark.sql.functions import count

df_by_year = (
    df.groupBy("occ_year")
      .agg(count("*").alias("total_crimes"))
      .orderBy("occ_year")
)

df_by_year.show()
```

```python
## Top 10 Crime Types
df_top_crimes = (
    df.groupBy("crm_cd_desc")
      .agg(count("*").alias("crime_count"))
      .orderBy("crime_count", ascending=False)
      .limit(10)
)

df_top_crimes.show()
```

```python
## Crime by Area_Name
df_by_area = (
    df.groupBy("area_name")
      .agg(count("*").alias("crime_count"))
      .orderBy("crime_count", ascending=False)
)

df_by_area.show()
```

```python
## Crimes by Age Group
df_by_age_group = (
    df.groupBy("age_group")
      .agg(count("*").alias("crime_count"))
      .orderBy("age_group")
)

df_by_age_group.show()
```

```python
%sql
SELECT
  occ_year,
  COUNT(*) AS total_crimes
FROM crime_final_cluster.default.clean_crime
GROUP BY occ_year
ORDER BY occ_year
```

```python
%sql
SELECT
  area_name,
  COUNT(*) AS total_crimes
FROM crime_final_cluster.default.clean_crime
GROUP BY area_name
ORDER BY total_crimes DESC
```
